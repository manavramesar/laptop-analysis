```python
ax = df['Brand'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Brands with the most laptops')
ax.set_ylabel('Count')
plt.show()
```

```python
ax = df.groupby('Brand')['Price'].mean().sort_values(ascending=True) \
    .plot(kind='barh') 
ax.set_xlabel('Average Price')
plt.show()
```
